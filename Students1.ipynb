```python
#Mean, Median, Standard Deviation, Variance for Math Score
mean_math = df['math score'].mean()
median_math = df['math score'].median()
sd = df['math score'].std()
var = df['math score'].var()
print("mean of math score is ", mean_math)
print("median of math score is",median_math)
print("Standard Deviation of math score is",sd)
print("Variance of math score is",var)
```

```python
# #Mean, Median, Standard Deviation, Variance for Reading Score
mean_math = df['reading score'].mean()
median_math = df['reading score'].median()
sd = df['reading score'].std()
var = df['reading score'].var()
print("mean of reading score is ", mean_math)
print("median of reading score is",median_math)
print("Standard Deviation of reading score is",sd)
print("Variance of reading score is",var)
```

```python
# #Mean, Median, Standard Deviation, Variance for Writing Score
mean_math = df['writing score'].mean()
median_math = df['writing score'].median()
sd = df['writing score'].std()
var = df['writing score'].var()
print("mean of writing score is ", mean_math)
print("median of writing score is",median_math)
print("Standard Deviation of writing score is",sd)
print("Variance of writing score is",var)
```
